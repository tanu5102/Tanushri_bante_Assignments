```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
numerical_columns = df.select_dtypes(include=['int64', 'float64'])

# Plot histograms for each numerical column
for column in numerical_columns.columns:
    plt.figure(figsize=(9, 9))
    sns.histplot(df[column], bins=30, kde=True)
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```

```python

import seaborn as sns

numerical_columns = df.select_dtypes(include=['int64', 'float64'])

# Analyze distribution characteristics
for column in numerical_columns.columns:
    print(f"Analysis for column: {column}")
    print(f"Skewness: {numerical_columns[column].skew()}")
    print(f"Presence of outliers: {True if numerical_columns[column].skew() > 1 or numerical_columns[column].skew() < -1 else False}")
    print("---------------------------------------------------")
```

```python
for column in numerical_columns.columns:
    plt.figure(figsize=(5, 5))
    sns.boxplot(x=df[column])
    plt.title(f'Boxplot of {column}')
    plt.xlabel(column)
    plt.show()
```

```python
for column in numerical_columns.columns:
    print(f"Analysis for column: {column}")
    q1 = numerical_columns[column].quantile(0.25)
    q3 = numerical_columns[column].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr

    outliers = numerical_columns[(numerical_columns[column] < lower_bound) | (numerical_columns[column] > upper_bound)]

    print(f"Interquartile range (IQR): {iqr}")
    print(f"Lower bound: {lower_bound}")
    print(f"Upper bound: {upper_bound}")
    print(f"Number of outliers: {len(outliers)}")
    print("---------------------------------------------------")
```

```python
categorical_columns = df.select_dtypes(include=['object'])
```

```python
for column in categorical_columns.columns:
    plt.figure(figsize=(5, 5))
    sns.countplot(data=df, x=column, order=df[column].value_counts().index)
    plt.title(f'Frequency of Categories in {column}')
    plt.xlabel(column)
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.show()
```

```python
for column in categorical_columns.columns:
    print(f"Analysis for column: {column}")
    category_counts = df[column].value_counts()
    num_categories = len(category_counts)
    most_common_category = category_counts.idxmax()
    num_most_common_category = category_counts.max()
    percent_most_common_category = (num_most_common_category / len(df)) * 100

    print(f"Number of categories: {num_categories}")
    print(f"Most common category: {most_common_category}")
    print(f"Number of occurrences of most common category: {num_most_common_category}")
    print(f"Percentage of most common category: {percent_most_common_category:.2f}%")
    print("---------------------------------------------------")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.boxplot(x='Volume', data=df)
plt.title('Boxplot of Volume')
plt.show()

sns.boxplot(x='Avg Price', data=df)
plt.title('Boxplot of Avg Price')
plt.show()

sns.boxplot(x='Total Sales Value', data=df)
plt.title('Boxplot of Total Sales Value')
plt.show()
```

```python
for TotalSalesValue in numerical_columns:
    mean = df[TotalSalesValue].mean()
    std_dev = df[TotalSalesValue].std()
    df[TotalSalesValue + '_standardized'] = (df[TotalSalesValue] - mean) / std_dev
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
df['Total Sales Value'].hist(bins=20, edgecolor='black')
plt.title('Before Standardization (Total Sales Value)')
plt.xlabel('Total Sales Value')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
df['Total Sales Value_standardized'].hist(bins=20, edgecolor='black')
plt.title('After Standardization (Sales)')
plt.xlabel('Total Sales Value (Standardized)')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
